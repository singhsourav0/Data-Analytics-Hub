```python
plt.figure(figsize=(10, 6))
sns.barplot(y='Number Of Ratings', x='Product Name', data=top10, palette='viridis')

# Add labels and title
plt.xlabel("Number of Ratings")
plt.ylabel("Product Name")
plt.title("Top 10 Rated Products")
plt.tight_layout()
plt.xticks(rotation = 60)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate average Sale Price by RAM
price_by_ram = data.groupby('Ram')['Sale Price'].mean().reset_index()

# Bar Plot: Sale Price vs RAM
plt.figure(figsize=(10, 6))
sns.barplot(x='Ram', y='Sale Price', data=price_by_ram, palette='coolwarm')
plt.title('Average Sale Price by RAM')
plt.xlabel('RAM')
plt.ylabel('Average Sale Price')
plt.tight_layout()
plt.show()
```

```python
# Aggregate average Sale Price by Star Rating
price_by_rating = data.groupby('Star Rating')['Sale Price'].mean().reset_index()

# Bar Plot: Sale Price vs Star Rating
plt.figure(figsize=(10, 6))
sns.barplot(x='Star Rating', y='Sale Price', data=price_by_rating, palette='viridis')
plt.title('Average Sale Price by Star Rating')
plt.xlabel('Star Rating')
plt.ylabel('Average Sale Price')
plt.tight_layout()
plt.show()
```

Which Brands Offer the Best Value for Money (Price and Ratings)?

```python
# Group data by Brand and calculate average Sale Price and Star Rating
value_brands = data.groupby('Brand').agg(
    avg_price=('Sale Price', 'mean'),
    avg_rating=('Star Rating', 'mean')
).reset_index()

# Sort by avg_rating descending and avg_price ascending for value
value_brands = value_brands.sort_values(by=['avg_rating', 'avg_price'], ascending=[False, True])
print(value_brands.head(10))  # Display top 10 value-for-money brands
```

Plot Sale Price vs MRP to Visualize Discounts

```python
# Plot Sale Price vs MRP as grouped bars
plt.figure(figsize=(15, 8))
sns.barplot(data=costly_phone.melt(id_vars=['Product Name'], value_vars=['Sale Price', 'Mrp']),
            x='Product Name', y='value', hue='variable')
plt.title('Sale Price vs MRP for Products')
plt.xticks(rotation=90, ha='right')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```
